```python
df.columns = ['ID', 'Delivery_person_ID', 'Delivery_person_Age',
       'Delivery_person_Ratings', 'Restaurant_latitude',
       'Restaurant_longitude', 'Delivery_location_latitude',
       'Delivery_location_longitude', 'Type_of_order', 'Type_of_vehicle',
       'Time_taken(min)']
df
```

```python
x = df.drop(columns=['Delivery_person_ID', 'ID'])
y = df['Time_taken(min)']
x
```

```python
y
```

```python
df['la'] = df['Restaurant_latitude'] - df['Delivery_location_latitude']
df['la'] = df['la'].abs()
df['la']
```

```python
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.rstrip()
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.replace('electric_scooter','scooter')
df['Type_of_order'] = df['Type_of_order'].str.replace('Buffet','Meal')
```

```python
df['Type_of_vehicle'].unique()
```

```python
vehicle_mapping = {'motorcycle': 2, 'scooter': 0, 'bicycle': 1}
x['Type_of_vehicle'] = x['Type_of_vehicle'].map(vehicle_mapping)
x
```

```python
order_mapping = {'Meal': 0, 'Drinks': 1, 'Snack': 2, 'Meal': 3}
x['Type_of_order'] = x['Type_of_order'].map(order_mapping)
x
```

```python
x
```

```python
df['Type_of_order'].unique()
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=5, random_state=42)
model.fit(train_X, train_y)
```

```python
model.score(train_X, train_y)
```

```python
model.score(test_X, test_y)
```
